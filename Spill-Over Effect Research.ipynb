```python
# Utility

# filters out 'advance'
def filter_advance(lst):
    return [*filter(lambda r: r != 'advance', lst)]

def assign_unit(virus_dict, response):
    """A function that will match a participant's response to an idea unit
    
    Args:
        param1 (dict): the virus dictionary
        param2 (str): the response given by the participant
        
    Returns:
        integer
    
    """
    val_list = list(virus_dict.values())
    key_list = list(virus_dict.keys())
    
    for i in val_list:
        num = jaccard_dist(i, response)
        
        if num > 0.25:
            index = val_list.index(i)
            return index + 1

# Helper function for jaccard_dist
def word_set(string):
    return set(string.split())


def jaccard_dist(str1, str2):
    """A function that computes the Jaccard distance of two strings

    Args:
        param1 (str): The first string.
        param2 (str): The second string.

    Returns:
        float: 
        
    """
    str1_set = word_set(str1)
    str2_set = word_set(str2)
    c = str1_set.intersection(str2_set)
    return float(len(c)) / (len(str1_set) + len(str2_set) - len(c))
```
